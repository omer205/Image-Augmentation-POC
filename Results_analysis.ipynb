## Metrics visualization per augmentation scheme

## Per Class Metrics

```python
df_preds = pd.read_csv('preds.csv', index_col=0)
df_preds.drop([col for col in df_preds.columns if col.startswith('ADC')], axis=1, inplace=True)
df_preds = df_preds[[col for col in df_preds.columns if col[-1].isdigit()]]
df_preds.head()
```

```python
y_true = pd.read_csv('true_labels.csv', index_col = 0)
y_true.columns =['true']
y_true
```

```python
preds = pd.concat([df_preds, y_true], axis=1)
```

```python
class_names = ['0-Fish', '1-Dog', '2-Soundsystem', '3-Chainsaw', '4-Cathedral', '5-French Horn', '6-Truck', '7-Gas Station', '8-Golf', '9-Parachute']
```

```python
from sklearn.metrics import precision_recall_fscore_support
prf_dict = {}
for col in df_preds.columns:
    p, r, f, _ = precision_recall_fscore_support(y_true, preds[col])
    prf_dict[col] = pd.DataFrame(dict([('precision', p),('recall', r), ('f1', f)]))
df_per_class = pd.concat(prf_dict, axis=0).reset_index().rename({'level_0':'run', 'level_1':'Class'}, axis=1)
```

```python
cols = ['precision', 'recall', 'f1']
df_per_class[cols] = 100*df_per_class[cols]
```

```python
df_per_class
```

```python
# palette = sns.color_palette("Paired", 5)
aug_dict = {
    'no_augs': 'None',
    'randaugment': 'RandAugment - M=7',
    'basic': 'Basic',
    'advanced': 'Basic - Added Augs',
    'randaugment_9_': 'RandAugment - M=9'
}

df_per_class['Augmentation'] = df_per_class['run'].apply(lambda s: s[:-1]).map(aug_dict)
data = df_per_class.groupby(['Augmentation', 'Class']).mean().reset_index()
```

```python
metrics = ['recall', 'precision', 'f1']
for metric in metrics:
    plt.figure(figsize=(25, 4))
    ax = sns.barplot(x='Class', 
                     y=metric, 
                     hue='Augmentation', 
                     data=data, palette=palette, 
                     alpha=0.8,
                     hue_order=['None', 'Basic', 'Basic - Added Augs',   'RandAugment - M=7',
                                'RandAugment - M=9']
                     )
    for container in ax.containers:
        ax.bar_label(container, fmt='%.0f%%', fontsize=10)
    ax.set_ylabel(f'{metric.capitalize()}[%]', fontsize=14)
    ax.set_xlabel('Class[%]', fontsize=14)

    ax.axvline(x=4.5, ymin=0, ymax=1, color='k')
    ax.set_xticklabels(class_names)
    ax.text(1.5, -17, 'Major Classes', fontdict={'fontsize': 14})
    ax.text(6.8, -17, 'Minor Classes', fontdict={'fontsize': 14})
    plt.legend(loc='lower right')
    plt.title(f'Class {metric.capitalize()}[%] Comparison (5 Folds average)', fontsize=14)
    plt.ylim(0, 102)
```

## Training History

```python
import pickle
with open('rgb_history_all.pkl', 'rb') as pickle_file:
    history = pickle.load(pickle_file)
```

```python
no_augs_history = pd.DataFrame.from_dict(history['no_augs'][2]).reset_index()
no_augs_history['Augmentation'] = 'None'
simple_history = pd.DataFrame.from_dict(history['ADC'][2]).reset_index()
simple_history['Augmentation'] = 'Simple Augmentation'
rand_history = pd.DataFrame.from_dict(history['randaugment'][2]).reset_index()
rand_history['Augmentation'] = 'RandAugment'
basic_history = pd.DataFrame.from_dict(history['basic'][2]).reset_index()
basic_history['Augmentation'] = 'basic'
advanced_history = pd.DataFrame.from_dict(history['advanced'][2]).reset_index()
advanced_history['Augmentation'] = 'Advanced Augmentation'
rand_9_history = pd.DataFrame.from_dict(history['randaugment_9'][2]).reset_index()
rand_9_history['Augmentation'] = 'RandAugment_9'
all_history = pd.concat([no_augs_history, basic_history, advanced_history,  rand_history,   rand_9_history])
all_history[['val_accuracy', 'accuracy']] = 100*all_history[['val_accuracy', 'accuracy']]
```

```python
plt.figure(figsize=(20,10))

ax = sns.lineplot(x='index', y='val_accuracy', hue='Augmentation', data=all_history, palette=palette, style='Augmentation',
            markers=['*']*5, linewidth=3, markersize=15)
sns.lineplot(x='index', y='accuracy', hue='Augmentation', data=all_history, palette=palette,style='Augmentation',
            markers=['o']*5, linewidth=3, markersize=12)
ax.set_xlabel('Epochs',fontsize=14)
ax.set_ylabel('Accuracy [%]', fontsize=14)
handles, labels = ax.get_legend_handles_labels()
labels = ['Val - None', 'Val - Basic', 'Val - Added Augs', 'Val - RandAugment M=7',   'Val - RandAugment M=9', 'Train - None',
          'Train - Basic', 'Train - Added Augs',  'Train - RandAugment M=7',  'Train - RandAugment M=9']
ax.legend(handles=handles, labels=labels, fontsize=14)
plt.title('Training History - Accuracy', fontdict={'size':20})
plt.yticks(np.arange(0, 100+1, 10))
plt.grid()
```

## 2D Visualizations of final dense features

```python
with open('test_rgb_high_dimension_rep.pkl', 'rb') as pickle_file:
    test_high_d_results = pickle.load(pickle_file)
pickle_file.close()

with open('train_rgb_high_dimension_rep.pkl', 'rb') as pickle_file:
    train_high_d_results = pickle.load(pickle_file)
pickle_file.close()
```

```python
aug_types = ['no_augs', 'randaugment']
test_high_d_dict = {aug: pd.DataFrame(test_high_d_results[aug]) for aug in aug_types}
train_high_d_dict = {aug: pd.DataFrame(train_high_d_results[aug]) for aug in aug_types}
y_true_test = y_true
y_true_train = pd.read_csv('train_true_labels.csv',index_col=0)
y_true_train
```

```python
from sklearn.manifold import TSNE, MDS
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.preprocessing import StandardScaler, MinMaxScaler
classes = [7,2]
dnn_reps_test = test_high_d_dict['randaugment']
sc = MinMaxScaler()

idx = np.where((y_true_test == classes[0]) | (y_true_test == classes[1]))[0]
dnn_reps_test = dnn_reps_test.loc[idx]
dnn_reps_test = sc.fit_transform(pd.DataFrame(dnn_reps_test))
y_true_filtered_test = y_true_test.loc[idx]

mds = MDS(max_iter=300, random_state=42)
reps_mds = mds.fit_transform(dnn_reps_test)
reps_df_test = pd.DataFrame([reps_mds[:,0], reps_mds[:,1], y_true_filtered_test.squeeze()]).T
reps_df_test.columns = ['0', '1', 'y_true']
reps_df_test['y_true'] = reps_df_test['y_true'].astype(int)
```

```python
dnn_reps_train = train_high_d_dict['randaugment']
idx = np.where((y_true_train == classes[0]) | (y_true_train == classes[1]))[0]
dnn_reps_train = dnn_reps_train.loc[idx]
dnn_reps_train = sc.fit_transform(pd.DataFrame(dnn_reps_train))
y_true_filtered_train = y_true_train.loc[idx]

mds = MDS(max_iter=300, random_state=42)
reps_mds = mds.fit_transform(dnn_reps_train)
reps_df_train = pd.DataFrame([reps_mds[:,0], reps_mds[:,1], y_true_filtered_train.squeeze()]).T
reps_df_train.columns = ['0', '1', 'y_true']
reps_df_train['y_true'] = reps_df_train['y_true'].astype(int)
```

```python
types = ['randaugment', 'no_augs']
title_dict = {'randaugment': 'RandAugment', 'no_augs': 'No Augmentations'}
classes = [2, 7]
for runtype in types:
    dnn_reps_test = test_high_d_dict[runtype]
    dnn_reps_train = train_high_d_dict[runtype]
    dnn_reps_test['pred'] = y_true_test
    dnn_reps_train['pred'] = y_true_train
    dnn_reps_test['set'] = 'test'
    dnn_reps_train['set'] = 'train'
    dnn_reps_train = dnn_reps_train.sample(frac=0.3)
    data = pd.concat([dnn_reps_test, dnn_reps_train])
    data = data[(data['pred'].isin(classes))]
    X = data.loc[:, (data.columns != 'pred') & (data.columns != 'set')]
    hue = data['pred']
    style = data['set']
    mds = MDS(max_iter=300, random_state=42)
    X_scaled = sc.fit_transform(X)
    reps_mds = mds.fit_transform(X_scaled)
    reps_df_test = pd.DataFrame([reps_mds[:, 0], reps_mds[:, 1], hue, style]).T
    reps_df_test.columns = ['0', '1', 'y_true', 'style']
    reps_df_test['y_true'] = reps_df_test['y_true'].astype(int)
    plt.figure(figsize=(10, 10))
    plt.xlim([-10, 10])
    plt.ylim([-10, 10])
    hue = reps_df_test[['y_true', 'style']].apply(tuple, axis=1)
    plt.title(
        f'{title_dict[runtype]} 2D Representation of Model Output\nClasses {classes}', fontsize=14)
    ax = sns.scatterplot(x='0', y='1', hue=hue,
                         data=reps_df_test,  alpha=0.6, palette=['g', 'lightgreen','b', 'skyblue'], s=70,
                         hue_order=sorted(hue.unique()))
    plt.legend(fontsize=12)
```
